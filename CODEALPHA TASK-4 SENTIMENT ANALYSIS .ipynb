### Step 2: Import Libraries and Load Data

```python
import pandas as pd
from textblob import TextBlob

# Load CSV
df = pd.read_csv('Web_Scrapping_Task.csv')
df.head()
```

### Step 3: Define a Function to Get Sentiment

```python
def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

### Step 4: Applying Sentiment Analysis

```python
df['Sentiment'] = df['Cleaned_Text'].apply(get_sentiment)
df.head()
```

### Step 5: Saving the Result to a new file

```python
df.to_csv('Cruz_Tech_Sentiment_Analysis_Results.csv', index=False)
```
